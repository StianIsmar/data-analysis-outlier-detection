### Plotting the data over cycles

```python
for i in range(1,21):
    string = "Sensor measurement " + str(i)
    cycles = df["Time, in cycles"]
    sensor = df[string]
    colors = (0,0,0)

    plt.scatter(cycles, sensor,linewidths=0.2)
    plt.figure(figsize=(16,8))
```
